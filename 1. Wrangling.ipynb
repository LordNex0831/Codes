##### Now, with all the steps above, we get a general idea about the data that we are dealing with, like the number of rows and columns in the data, the data types of each feature, and total missing features.

##### Now many missing values are marked as '?' in the fields, so first we will replace those with'NaN' values

```python
# Replacing '?' with 'NaN' values
df.replace("?", np.nan, inplace = True)
```

```python
# Generating a list of missing values for each column
col_list=df.columns.tolist()

for i in col_list:
    print(df[i].isna().value_counts())
```

##### Now there are quite a few fields with missing values. So, before replacing the missing values, we will remove fields that we will not require.

```python
# printing a list of columns
df.columns
```

```python
# Dropping column fields
df=df.drop(['SEVERITYCODE.1','COLDETKEY','STATUS','INTKEY','INCDTTM','OBJECTID',
            'INCKEY','EXCEPTRSNCODE','EXCEPTRSNDESC','REPORTNO','SEVERITYDESC','SDOT_COLDESC',
            'SDOTCOLNUM','ST_COLDESC','SEGLANEKEY','CROSSWALKKEY','ST_COLCODE','SDOT_COLCODE'],axis=1)
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

##### We dropped some column fields from the original dataset, so now we are good to proceed with replacing the missing data values

```python
# Generating a summary of missing values for the dataset
col_list=df.columns.tolist()

for i in col_list:
    if(df[i].isna().sum()>0):  # Condition to print only columns fields with missing values
        print('Total missing values in field '+i+':',df[i].isna().sum())
        print(df[i].value_counts())
        print('----------------')
```

```python
# Now we will generate a summary of fields with no missing values
for i in col_list:
    if(df[i].isna().sum()==0):
         print(df[i].value_counts())
         print('--------x--------')
```

```python
# Replacing missing data fields with suitable values
df['ADDRTYPE'].replace(np.nan,'Not Specified',inplace=True)
df['LOCATION'].replace(np.nan,'Not Specified',inplace=True)
df['COLLISIONTYPE'].replace(np.nan,'Other',inplace=True)
df['JUNCTIONTYPE'].replace(np.nan,'Unknown',inplace=True)
df['INATTENTIONIND'].replace(np.nan,'N',inplace=True)
df['PEDROWNOTGRNT'].replace(np.nan,'N',inplace=True)
df['SPEEDING'].replace(np.nan,'N',inplace=True)
df['LIGHTCOND'].replace(np.nan,'Unknown',inplace=True)
df['LIGHTCOND'].replace('Other','Unknown',inplace=True)
df['ROADCOND'].replace(np.nan,'Unknown',inplace=True)
df['ROADCOND'].replace('Other','Unknown',inplace=True)
df['WEATHER'].replace(np.nan,'Unknown',inplace=True)
df['WEATHER'].replace('Other','Unknown',inplace=True)
df['UNDERINFL'].replace(np.nan,'N',inplace=True)
df['UNDERINFL'].replace('0','N',inplace=True)
df['UNDERINFL'].replace('1','Y',inplace=True)
```

```python
# checking number of missing fields
for i in col_list:
    if(df[i].isna().sum()>0):
         print('Total missing values in field '+i+':',df[i].isna().sum())
         print(df[i].value_counts())
         print('xxxxxxxxxxxxxxxxx')
```

```python
# now checking fields with no missing values
for i in col_list:
    if(df[i].isna().sum()==0):
         print(df[i].value_counts())
         print('--------x--------')
```
